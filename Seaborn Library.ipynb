```python
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
tip=sns.load_dataset('tips')
```

```python
tip.head()
```

```python
#Correlation with heatmap
tip.corr()
```

```python
sns.heatmap(tip.corr())
```

### JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot,  a 2D histogram or a 2D density plot

### Univariate Analysis

```python
sns.jointplot(x='total_bill', y='tip', data=tip, kind='hex')
```

### Pair Plot

A "pair plot" is also known as a scatterplot, in which one variable in the same data row is matched with anaother variable's value, like this. Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(tip)
```

```python
sns.pairplot(tip, hue='smoker')
```

### Dist Plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.displot(tip['tip'])
```

```python
sns.displot(tip['tip'], kde=False, bins=10)
```

### Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points.

1. Box Plot
2. Violin Plot
3. Count Plot
4. Bar Plot

```python
#Count Plot 
sns.countplot(x=tip['day'])
```

```python
sns.countplot(y=tip['sex'])
```

```python
sns.barplot(x=tip['total_bill'], y=tip['sex'])
```

```python
sns.barplot(x=tip['sex'], y=tip['total_bill'])
```

### Box Plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a 5-number summary

```python
sns.boxplot(x=tip['sex'], y=tip['total_bill'])
```

```python
sns.boxplot(x=tip['day'], y=tip['total_bill'])
```

```python
sns.boxplot(data=tip, orient='v')
```

```python
sns.boxplot(x=tip['total_bill'], y=tip['day'])
```

```python
sns.boxplot(x=tip['total_bill'], y=tip['day'], hue=tip['sex'])
```

### Violin Plot
Violin Plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x=tip['total_bill'], y=tip['day'], palette='rainbow')
```
